```python
def plot_best_fit(kd, Pkd, cov, kf, Pkf, save=None, color="blue"):
    plt.errorbar(kd, Pkd, yerr=np.sqrt(cov), ls='none', marker='.', label="data", color=color)
    plt.loglog(kf, Pkf, linestyle = 'dashdot', label="bestfit", color=color)
    
    plt.legend(frameon=False)
    plt.xlabel('$k$')
    plt.ylabel('$P(k)$')
    plt.tight_layout()
    
    if save is not None:
        plt.savefig(save, bbox_inches='tight')
    plt.show()
    
def plot_comparison(kdfkp, Pkdfkp, covfkp, kffkp, Pkffkp, kd, Pkd, cov, kf, Pkf, save=None, colorfkp="orange", color="blue"):
    
    plt.errorbar(kdfkp, Pkdfkp, yerr=np.sqrt(covfkp), ls='none', marker='s', markersize=3, label="FKP", color=colorfkp)
    plt.loglog(kffkp, Pkffkp, linestyle = 'dashed', label="FKP", color=colorfkp)
    
    plt.errorbar(kd, Pkd, yerr=np.sqrt(cov), ls='none', marker='.', label="Optimal", color=color)
    plt.loglog(kf, Pkf, linestyle = 'dashdot', label="Optimal", color=color)
    
    plt.legend(frameon=False)
    plt.xlabel('$k$')
    plt.ylabel('$P(k)$')
    plt.tight_layout()
    
    if save is not None:
        plt.savefig(save, bbox_inches='tight')
    plt.show()
```

```python
def chi2_dof(data, model, cov):
    dof = len(data) - 1
    chi2 = np.sum((data - model)**2/cov)
    
    return chi2/dof
```

```python
def plot_ax(ax, kd, Pkd, kf, Pkf, cov, linestyle, marker, label, color, annotate=None, xy=(20,20)):
    ax.errorbar(kd, Pkd, yerr=np.sqrt(cov), 
                 ls='none', marker=marker, markersize=4, label=label, color=color)
    ax.loglog(kf, Pkf, 
               linestyle = linestyle, color=color)
    
    ax.legend(loc=1,frameon=False)
    ax.set_xlabel('$k$ [$h$/Mpc]')
    ax.set_ylabel('$P_0(k)$')
    
    if annotate is not None:
        ax.annotate(annotate, xy=xy, xycoords='axes pixels')
```
